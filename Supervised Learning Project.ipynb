```python
"""
Bank Customer Churn Prediction - Supervised Learning Project
=============================================================
Complete Machine Learning Project
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, roc_auc_score, classification_report, confusion_matrix, roc_curve, precision_score, recall_score
from sklearn.utils import resample
import warnings
warnings.filterwarnings('ignore')

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
# ==================================================================
# PROJECT INTRODUCTION
# ==================================================================

"""
PROJECT: BANK CUSTOMER CHURN PREDICTION

BUSINESS CONTEXT:
Beta Bank is experiencing a gradual loss of customers each month. The bank has 
identified that retaining existing customers is more cost-effective than acquiring 
new ones. It is necessary to predict whether a customer will leave the bank in the 
near future in order to take preventive measures.

MAIN OBJECTIVE:
Develop a machine learning model that accurately predicts whether a customer will 
leave the bank, achieving a minimum F1-score of 0.59.

SPECIFIC OBJECTIVES:
1. Analyze the dataset and prepare features for modeling
2. Implement and evaluate multiple techniques to handle class imbalance:
   - Upsampling
   - Downsampling
   - Class weight adjustment
3. Train and compare different classification algorithms:
   - Random Forest
   - Logistic Regression
   - Decision Trees
4. Evaluate models using appropriate metrics (F1-score and AUC-ROC)
5. Select the best model and validate its performance on test data

SUCCESS METRIC:
The model must achieve an F1-score of at least 0.59 on the test set,
properly balancing precision and recall in identifying customers who 
will potentially leave the bank.

WORK PLAN:
1. Data loading and initial exploration
2. Data preprocessing and splitting
3. Model training with different balancing techniques
4. Model evaluation and comparison
5. Final testing of the best model
6. Results analysis and conclusions
"""
print("="*80)
print("BANK CUSTOMER CHURN PREDICTION - SUPERVISED LEARNING PROJECT")
print("="*80)
print("\nğŸ“Š Loading and preparing data...\n")
```

```python
df = pd.read_csv('/datasets/Churn.csv')

print(f"âœ… Dataset loaded successfully!")
print(f"  Shape: {df.shape}")
print(f"  Rows: {df.shape[0]:,}")
print(f"  Columns: {df.shape[1]}")

print(f"\nğŸ” Missing Values Check:")
missing = df.isnull().sum()
if missing.sum() > 0:
    print(missing[missing > 0])
    print(f"\n  âš ï¸ Found {missing.sum()} missing values")
else:
    print("  âœ“ No missing values found")

duplicates = df.duplicated().sum()
print(f"\nğŸ” Duplicate Rows: {duplicates}")
```

```python
print(f"\n{'='*80}")
print("DATA CLEANING AND PREPROCESSING")
print("="*80)

columns_to_drop = ['RowNumber', 'CustomerId', 'Surname']
df_clean = df.drop(columns=columns_to_drop)

print(f"\nâœ“ Dropped columns: {columns_to_drop}")
print(f"âœ“ Remaining columns: {list(df_clean.columns)}")

print(f"\nğŸ”§ Handling missing values...")
for col in df_clean.columns:
    if df_clean[col].isnull().sum() > 0:
        if df_clean[col].dtype in ['float64', 'int64']:
            df_clean[col].fillna(df_clean[col].median(), inplace=True)
            print(f"  âœ“ Filled {col} with median")
        else:
            df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
            print(f"  âœ“ Filled {col} with mode")

print(f"âœ“ All missing values handled")
print(f"âœ“ Final dataset shape: {df_clean.shape}")
```

```python

print(f"\n{'='*80}")
print("CLASS BALANCE ANALYSIS")
print("="*80)

class_counts = df_clean['Exited'].value_counts()
class_percentages = df_clean['Exited'].value_counts(normalize=True) * 100

print(f"\nğŸ“Š Target Variable Distribution:")
print(f"  Class 0 (Stayed): {class_counts[0]:,} ({class_percentages[0]:.2f}%)")
print(f"  Class 1 (Left): {class_counts[1]:,} ({class_percentages[1]:.2f}%)")

imbalance_ratio = class_counts[0] / class_counts[1]
print(f"\nâš ï¸ Imbalance Ratio: {imbalance_ratio:.2f}:1")
print(f"   The dataset is IMBALANCED!")
print(f"   Majority class is {imbalance_ratio:.2f}x larger than minority class")
```

```python
print(f"\n{'='*80}")
print("FEATURE ENGINEERING")
print("="*80)

X = df_clean.drop('Exited', axis=1)
y = df_clean['Exited']

print(f"\nâœ“ Features shape: {X.shape}")
print(f"âœ“ Target shape: {y.shape}")

print(f"\nğŸ”§ Encoding categorical variables...")
X['Geography'] = X['Geography'].map({'France': 0, 'Spain': 1, 'Germany': 2})
X['Gender'] = X['Gender'].map({'Female': 0, 'Male': 1})

print(f"âœ“ Geography encoded: {X['Geography'].unique()}")
print(f"âœ“ Gender encoded: {X['Gender'].unique()}")
```

```python
# ==================================================================
# FEATURE ENGINEERING
# ==================================================================

print(f"\n{'='*80}")
print("FEATURE ENGINEERING")
print("="*80)

print(f"\nğŸ“Š Separating features and target variable...")
X = df_clean.drop('Exited', axis=1)
y = df_clean['Exited']

print(f"âœ“ Features shape: {X.shape}")
print(f"âœ“ Target shape: {y.shape}")

# One-Hot Encoding for categorical variables
print(f"\nğŸ”§ Applying One-Hot Encoding to categorical variables...")
categorical_columns = ['Geography', 'Gender']

print(f"   Categorical columns found: {categorical_columns}")

X_encoded = pd.get_dummies(X, columns=categorical_columns, drop_first=True)

print(f"âœ“ One-Hot Encoding completed")
print(f"âœ“ New feature shape: {X_encoded.shape}")
print(f"âœ“ Feature columns: {list(X_encoded.columns)}")

X = X_encoded

print(f"\nâœ… Feature engineering completed!")
print(f"   Final features: {X.shape[1]} columns")
print(f"\n{'='*80}")
print("DATA SPLITTING")
print("="*80)

X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full, test_size=0.25, random_state=RANDOM_STATE, stratify=y_train_full
)

print(f"\nâœ“ Training set: {X_train.shape[0]:,} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"âœ“ Validation set: {X_valid.shape[0]:,} samples ({X_valid.shape[0]/len(X)*100:.1f}%)")
print(f"âœ“ Test set: {X_test.shape[0]:,} samples ({X_test.shape[0]/len(X)*100:.1f}%)")

print(f"\nğŸ”§ Scaling features...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled = scaler.transform(X_test)

print(f"âœ“ Features scaled using StandardScaler")
```

```python
print(f"\n{'='*80}")
print("BASELINE MODELS (Without Imbalance Correction)")
print("="*80)

def evaluate_model(model, X_val, y_val, model_name="Model"):
    """Evaluate model and return metrics"""
    y_pred = model.predict(X_val)
    y_pred_proba = model.predict_proba(X_val)[:, 1]
    
    f1 = f1_score(y_val, y_pred)
    auc_roc = roc_auc_score(y_val, y_pred_proba)
    precision = precision_score(y_val, y_pred)
    recall = recall_score(y_val, y_pred)
    
    print(f"\n{'â”€'*60}")
    print(f"{model_name}")
    print(f"{'â”€'*60}")
    print(f"  F1 Score: {f1:.4f}")
    print(f"  AUC-ROC: {auc_roc:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    
    return f1, auc_roc, precision, recall

print(f"\nğŸ¤– Training Random Forest Baseline...")
rf_baseline = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf_baseline.fit(X_train_scaled, y_train)
f1_rf_base, auc_rf_base, prec_rf_base, rec_rf_base = evaluate_model(
    rf_baseline, X_valid_scaled, y_valid, "Random Forest Baseline"
)

print(f"\nğŸ¤– Training Logistic Regression Baseline...")
lr_baseline = LogisticRegression(
    max_iter=1000,
    random_state=RANDOM_STATE
)
lr_baseline.fit(X_train_scaled, y_train)
f1_lr_base, auc_lr_base, prec_lr_base, rec_lr_base = evaluate_model(
    lr_baseline, X_valid_scaled, y_valid, "Logistic Regression Baseline"
)

print(f"\nğŸ“ Baseline Findings:")
print(f"  âš ï¸ Models without imbalance correction show poor minority class performance")
print(f"  âš ï¸ Need to apply techniques to improve F1 score")
# ==================================================================
# HYPERPARAMETER TUNING FOR RANDOM FOREST
# ==================================================================

print(f"\n{'='*80}")
print("HYPERPARAMETER TUNING - RANDOM FOREST")
print("="*80)
print("\nğŸ”§ Testing different max_depth values to find optimal configuration...")

max_depth_values = [2, 4, 6, 8, 10, 12, 15, 20, None]
tuning_results = []

print(f"\n{'Depth':<10} {'F1 Score':<12} {'AUC-ROC':<12} {'Precision':<12} {'Recall':<12}")
print("â”€" * 60)

for depth in max_depth_values:
    rf_temp = RandomForestClassifier(
        n_estimators=100,
        max_depth=depth,
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    rf_temp.fit(X_train_scaled, y_train)
    
    y_pred = rf_temp.predict(X_valid_scaled)
    y_pred_proba = rf_temp.predict_proba(X_valid_scaled)[:, 1]
    
    f1 = f1_score(y_valid, y_pred)
    auc = roc_auc_score(y_valid, y_pred_proba)
    precision = precision_score(y_valid, y_pred)
    recall = recall_score(y_valid, y_pred)
    
    tuning_results.append({
        'max_depth': depth if depth is not None else 'None',
        'f1_score': f1,
        'auc_roc': auc,
        'precision': precision,
        'recall': recall
    })
    
    depth_str = str(depth) if depth is not None else "None"
    print(f"{depth_str:<10} {f1:<12.4f} {auc:<12.4f} {precision:<12.4f} {recall:<12.4f}")

best_result = max(tuning_results, key=lambda x: x['f1_score'])
best_depth = best_result['max_depth']
best_f1 = best_result['f1_score']

print(f"\n{'='*60}")
print(f"âœ… BEST CONFIGURATION FOUND:")
print(f"   max_depth = {best_depth}")
print(f"   F1 Score = {best_f1:.4f}")
print(f"   AUC-ROC = {best_result['auc_roc']:.4f}")
print(f"{'='*60}")

print(f"\nğŸ¤– Training Random Forest with optimized max_depth={best_depth}...")
best_depth_value = None if best_depth == 'None' else best_depth

rf_tuned = RandomForestClassifier(
    n_estimators=100,
    max_depth=best_depth_value,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf_tuned.fit(X_train_scaled, y_train)

print(f"âœ… Tuned Random Forest model trained successfully!")
print(f"\nğŸ“Š Tuning Conclusion:")
print(f"   The optimal max_depth of {best_depth} provides the best balance")
print(f"   between model complexity and generalization performance.")
```

```python
print(f"\n{'='*80}")
print("TECHNIQUE 1: UPSAMPLING")
print("="*80)

X_train_df = pd.DataFrame(X_train_scaled, columns=X.columns)
X_train_df['Exited'] = y_train.values

df_majority = X_train_df[X_train_df['Exited'] == 0]
df_minority = X_train_df[X_train_df['Exited'] == 1]

print(f"\nğŸ“Š Before upsampling:")
print(f"  Majority: {len(df_majority):,}")
print(f"  Minority: {len(df_minority):,}")

df_minority_upsampled = resample(
    df_minority,
    replace=True,
    n_samples=len(df_majority),
    random_state=RANDOM_STATE
)

df_upsampled = pd.concat([df_majority, df_minority_upsampled])
X_train_upsampled = df_upsampled.drop('Exited', axis=1).values
y_train_upsampled = df_upsampled['Exited'].values

print(f"\nâœ“ After upsampling:")
print(f"  Total: {len(X_train_upsampled):,}")
print(f"  Class 0: {np.sum(y_train_upsampled == 0):,}")
print(f"  Class 1: {np.sum(y_train_upsampled == 1):,}")
print(f"  âœ“ Classes balanced!")

print(f"\nğŸ¤– Training Random Forest with Upsampling...")
rf_upsample = RandomForestClassifier(
    n_estimators=150,
    max_depth=12,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf_upsample.fit(X_train_upsampled, y_train_upsampled)
f1_rf_up, auc_rf_up, prec_rf_up, rec_rf_up = evaluate_model(
    rf_upsample, X_valid_scaled, y_valid, "Random Forest + Upsampling"
)

print(f"\nğŸ¤– Training Logistic Regression with Upsampling...")
lr_upsample = LogisticRegression(
    max_iter=1000,
    random_state=RANDOM_STATE
)
lr_upsample.fit(X_train_upsampled, y_train_upsampled)
f1_lr_up, auc_lr_up, prec_lr_up, rec_lr_up = evaluate_model(
    lr_upsample, X_valid_scaled, y_valid, "Logistic Regression + Upsampling"
)
# ==================================================================
# HYPERPARAMETER TUNING - RANDOM FOREST WITH UPSAMPLING
# ==================================================================

print(f"\n{'='*80}")
print("HYPERPARAMETER TUNING - RANDOM FOREST WITH UPSAMPLING")
print("="*80)
print("\nğŸ”§ Testing different max_depth values with upsampled data...")

max_depth_values = [2, 4, 6, 8, 10, 12, 15, 20, None]
tuning_results_upsample = []

print(f"\n{'Depth':<10} {'F1 Score':<12} {'AUC-ROC':<12} {'Precision':<12} {'Recall':<12}")
print("â”€" * 60)

for depth in max_depth_values:
    rf_temp = RandomForestClassifier(
        n_estimators=150,
        max_depth=depth,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    rf_temp.fit(X_train_upsampled, y_train_upsampled)
    
    y_pred = rf_temp.predict(X_valid_scaled)
    y_pred_proba = rf_temp.predict_proba(X_valid_scaled)[:, 1]
    
    f1 = f1_score(y_valid, y_pred)
    auc = roc_auc_score(y_valid, y_pred_proba)
    precision = precision_score(y_valid, y_pred)
    recall = recall_score(y_valid, y_pred)
    
    tuning_results_upsample.append({
        'max_depth': depth if depth is not None else 'None',
        'f1_score': f1,
        'auc_roc': auc,
        'precision': precision,
        'recall': recall
    })
    
    depth_str = str(depth) if depth is not None else "None"
    print(f"{depth_str:<10} {f1:<12.4f} {auc:<12.4f} {precision:<12.4f} {recall:<12.4f}")

best_result_up = max(tuning_results_upsample, key=lambda x: x['f1_score'])
best_depth_up = best_result_up['max_depth']
best_f1_up = best_result_up['f1_score']

print(f"\n{'='*60}")
print(f"âœ… BEST CONFIGURATION WITH UPSAMPLING:")
print(f"   max_depth = {best_depth_up}")
print(f"   F1 Score = {best_f1_up:.4f}")
print(f"   AUC-ROC = {best_result_up['auc_roc']:.4f}")
print(f"   Precision = {best_result_up['precision']:.4f}")
print(f"   Recall = {best_result_up['recall']:.4f}")
print(f"{'='*60}")

print(f"\nğŸ¤– Training optimized Random Forest with upsampling (max_depth={best_depth_up})...")
best_depth_value_up = None if best_depth_up == 'None' else best_depth_up

rf_upsample_tuned = RandomForestClassifier(
    n_estimators=150,
    max_depth=best_depth_value_up,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf_upsample_tuned.fit(X_train_upsampled, y_train_upsampled)

f1_rf_up_tuned, auc_rf_up_tuned, prec_rf_up_tuned, rec_rf_up_tuned = evaluate_model(
    rf_upsample_tuned, X_valid_scaled, y_valid, "Random Forest + Upsampling (Tuned)"
)

print(f"\nâœ… Hyperparameter tuning with upsampling completed!")
print(f"   Improvement over default: {(best_f1_up - f1_rf_up)*100:.2f}% in F1-score")
```

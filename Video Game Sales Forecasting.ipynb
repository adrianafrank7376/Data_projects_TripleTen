```python
df.columns = df.columns.str.lower()

print("DATA EXPLORATION: Key Questions to Answer")
print("\n" + "="*80 + "\n")

print("QUESTION 1: What's the total number of records in our dataset?")
print("-"*80)

total_records = len(df)
total_columns = len(df.columns)

print(f"\nTotal number of records (rows): {total_records}")
print(f"Total number of columns: {total_columns}")
print(f"Dataset dimensions: {df.shape[0]} rows √ó {df.shape[1]} columns")

print("\nColumn names:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")

print("\nDataset size:")
memory_usage = df.memory_usage(deep=True).sum() / (1024 ** 2)
print(f"Memory usage: {memory_usage:.2f} MB")

print("\n" + "="*80 + "\n")

print("QUESTION 2: What data types are present in each column?")
print("-"*80)

print("\nData types by column:")
for col in df.columns:
    dtype = df[col].dtype
    non_null = df[col].notna().sum()
    null_count = df[col].isna().sum()
    unique_values = df[col].nunique()
    
    print(f"\n{col}:")
    print(f"  Data type: {dtype}")
    print(f"  Non-null values: {non_null}")
    print(f"  Null values: {null_count}")
    print(f"  Unique values: {unique_values}")
    
    if dtype == 'object':
        print(f"  Sample values: {df[col].dropna().unique()[:5]}")
    elif dtype in ['float64', 'int64']:
        print(f"  Range: {df[col].min()} to {df[col].max()}")

print("\n")
print("Summary of data types:")
dtype_counts = df.dtypes.value_counts()
print(dtype_counts)

print("\n")
print("Data type classification:")
print(f"Numerical columns: {len(df.select_dtypes(include=['float64', 'int64']).columns)}")
print(f"Categorical columns: {len(df.select_dtypes(include=['object']).columns)}")

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"\nNumerical: {numerical_cols}")
print(f"Categorical: {categorical_cols}")

print("\n" + "="*80 + "\n")

print("QUESTION 3: Are there any obvious issues with the data?")
print("-"*80)

print("\nISSUE 1: Missing Values")
print("-"*40)

missing_summary = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df) * 100).round(2)
}).sort_values('Missing_Percentage', ascending=False)

print("\nMissing values analysis:")
print(missing_summary)

print("\nSeverity assessment:")
for _, row in missing_summary.iterrows():
    if row['Missing_Percentage'] > 50:
        severity = "CRITICAL"
    elif row['Missing_Percentage'] > 20:
        severity = "MODERATE"
    elif row['Missing_Percentage'] > 0:
        severity = "MINOR"
    else:
        severity = "NONE"
    
    if row['Missing_Percentage'] > 0:
        print(f"{severity} - {row['Column']}: {row['Missing_Percentage']:.1f}% missing")

print("\n")
print("ISSUE 2: Data Type Problems")
print("-"*40)

print("\n'user_score' column issues:")
user_score_unique = df['user_score'].unique()
print(f"Unique values found: {len(user_score_unique)}")
print(f"Sample values: {user_score_unique[:15]}")

if 'tbd' in df['user_score'].values:
    tbd_count = (df['user_score'] == 'tbd').sum()
    print(f"Found 'tbd' (To Be Determined) values: {tbd_count}")
    print(f"   Percentage: {(tbd_count/len(df)*100):.2f}%")
    print(f"   Issue: Text values in numerical column")

print("\n'year_of_release' column issues:")
year_dtypes = df['year_of_release'].dtype
print(f"Current data type: {year_dtypes}")
if year_dtypes == 'object':
    print(f"Year stored as text instead of number")
    print(f"   Sample values: {df['year_of_release'].unique()[:10]}")

print("\n")
print("ISSUE 3: Duplicates")
print("-"*40)

duplicate_count = df.duplicated().sum()
print(f"\nTotal duplicate rows: {duplicate_count}")

if duplicate_count > 0:
    print("Duplicate records found")
    print(df[df.duplicated(keep=False)].head())
else:
    print("No complete duplicate rows found")

name_duplicates = df['name'].duplicated().sum()
print(f"\nDuplicate game names: {name_duplicates}")
if name_duplicates > 0:
    print("Note: Same game may appear on multiple platforms (expected)")
    print("\nExample - same game on different platforms:")
    sample_game = df[df['name'].duplicated(keep=False)]['name'].iloc[0]
    print(df[df['name'] == sample_game][['name', 'platform', 'year_of_release']].head())

print("\n")
print("ISSUE 4: Outliers in Sales Data")
print("-"*40)

sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']

print("\nSales outlier analysis:")
for col in sales_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_pct = (len(outliers) / len(df)) * 100
    
    print(f"\n{col}:")
    print(f"  Q1: ${q1:.2f}M, Q3: ${q3:.2f}M, IQR: ${iqr:.2f}M")
    print(f"  Outlier bounds: ${lower_bound:.2f}M to ${upper_bound:.2f}M")
    print(f"  Outliers found: {len(outliers)} ({outlier_pct:.1f}%)")
    
    if len(outliers) > 0:
        max_outlier = df[col].max()
        print(f"  Max value: ${max_outlier:.2f}M")
        print(f"  Note: High sales outliers are blockbuster games (not errors)")

print("\n")
print("ISSUE 5: Inconsistent or Invalid Values")
print("-"*40)

print("\nYear of Release validation:")
min_year = df['year_of_release'].min()
max_year = df['year_of_release'].max()
print(f"Year range: {min_year} to {max_year}")

invalid_years = df[(df['year_of_release'] < 1970) | (df['year_of_release'] > 2017)]
print(f"Potentially invalid years: {len(invalid_years)}")
if len(invalid_years) > 0:
    print("Note: Very old years may be data errors or prototype entries")

print("\nPlatform validation:")
platform_count = df['platform'].nunique()
print(f"Number of unique platforms: {platform_count}")
print(f"Platforms: {sorted(df['platform'].unique())}")

print("\nGenre validation:")
genre_count = df['genre'].nunique()
print(f"Number of unique genres: {genre_count}")
print(f"Genres: {sorted(df['genre'].dropna().unique())}")

print("\nRating validation:")
rating_count = df['rating'].nunique()
print(f"Number of unique ratings: {rating_count}")
print(f"Ratings: {sorted(df['rating'].dropna().unique())}")

print("\n")
print("ISSUE 6: Score Range Validation")
print("-"*40)

print("\nCritic Score (should be 0-100):")
critic_min = df['critic_score'].min()
critic_max = df['critic_score'].max()
print(f"Range: {critic_min} to {critic_max}")
if critic_min < 0 or critic_max > 100:
    print("Values outside expected range (0-100)")
else:
    print("Values within expected range")

print("\nUser Score (should be 0-10):")
user_numeric = pd.to_numeric(df['user_score'], errors='coerce')
user_min = user_numeric.min()
user_max = user_numeric.max()
print(f"Range: {user_min} to {user_max}")
if user_min < 0 or user_max > 10:
    print("Values outside expected range (0-10)")
else:
    print("Values within expected range")

print("\n" + "="*80 + "\n")

print("QUESTION 4: Do we see any immediate patterns or anomalies?")
print("-"*80)

print("\nPATTERN 1: Temporal Distribution")
print("-"*40)

games_by_decade = df.groupby((df['year_of_release'] // 10 * 10).astype('Int64'))['name'].count()
print("\nGames released by decade:")
print(games_by_decade)

print("\nObservation:")
print("  - Sparse data before 1990s")
print("  - Massive growth in 2000s")
print("  - Peak in 2000-2010 period")
print("  - Recent decline (may indicate incomplete 2016 data)")

print("\n")
print("PATTERN 2: Platform Popularity")
print("-"*40)

top_platforms = df['platform'].value_counts().head(10)
print("\nTop 10 platforms by number of games:")
print(top_platforms)

print("\nObservation:")
print("  - DS and PS2 have most games released")
print("  - Multiple PlayStation generations present")
print("  - PC gaming well-represented")

print("\n")
print("PATTERN 3: Genre Distribution")
print("-"*40)

genre_distribution = df['genre'].value_counts()
print("\nGames by genre:")
print(genre_distribution)

print("\nObservation:")
print("  - Action dominates with most releases")
print("  - Sports and Shooter genres popular")
print("  - Relatively balanced distribution across genres")

print("\n")
print("PATTERN 4: Regional Sales Patterns")
print("-"*40)

total_sales_by_region = pd.Series({
    'North America': df['na_sales'].sum(),
    'Europe': df['eu_sales'].sum(),
    'Japan': df['jp_sales'].sum(),
    'Other': df['other_sales'].sum()
})

print("\nTotal sales by region:")
print(total_sales_by_region)

print("\nMarket share:")
for region, sales in total_sales_by_region.items():
    percentage = (sales / total_sales_by_region.sum()) * 100
    print(f"  {region}: {percentage:.1f}%")

print("\nObservation:")
print("  - North America is largest market")
print("  - Europe second-largest")
print("  - Japan significant but smaller")
print("  - Regional preferences likely differ")

print("\n")
print("ANOMALY: 2016 Data Incompleteness")
print("-"*40)

data_2016 = df[df['year_of_release'] == 2016]
data_2015 = df[df['year_of_release'] == 2015]

print(f"\nGames released in 2015: {len(data_2015)}")
print(f"Games released in 2016: {len(data_2016)}")
print(f"Difference: {len(data_2015) - len(data_2016)} fewer games")

print("\nObservation:")
print("  2016 data is incomplete (as noted in project description)")
print("  - Fewer games recorded")
print("  - Should weight 2016 data carefully in analysis")

print("\n" + "="*80 + "\n")

print("SUMMARY OF DATA QUALITY ASSESSMENT")
print("="*80)

print("\nSTRENGTHS:")
print(f"  - Large dataset with {total_records} records")
print("  - Multiple years of historical data (1980-2016)")
print("  - Comprehensive regional coverage")
print("  - Sales data appears complete")
print("  - Multiple attributes for analysis")

print("\nISSUES IDENTIFIED:")
print("  1. High percentage of missing review scores (50%+)")
print("  2. Missing ESRB ratings (60%)")
print("  3. 'TBD' text values in user_score column")
print("  4. 2016 data incomplete")
print("  5. Some very old records with sparse data")

print("\nRECOMMENDED ACTIONS:")
print("  1. Convert year_of_release and user_score to numeric")
print("  2. Handle 'tbd' values as NaN")
print("  3. Focus analysis on 2013-2016 period")
print("  4. Use dropna() selectively per analysis need")
print("  5. Document limitations in conclusions")

print("\nDATA IS SUITABLE FOR ANALYSIS:")
print("Despite identified issues, the dataset contains sufficient quality data")
print("for meaningful analysis and forecasting. Issues are documented and will")
print("be handled appropriately in the data preparation phase.")

print("\n" + "="*80)
```

```python
print("PLATFORM LIFECYCLE ANALYSIS")
print("\n" + "="*80 + "\n")

print("QUESTION 1: Which platforms show consistent sales over time?")
print("-" * 80)

platform_yearly_sales = df.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()

consistent_platforms = []

for platform in df['platform'].unique():
    platform_data = platform_yearly_sales[platform_yearly_sales['platform'] == platform]
    
    if len(platform_data) >= 5:
        sales_values = platform_data['total_sales'].values
        years_active = len(platform_data)
        total_sales = sales_values.sum()
        avg_sales = sales_values.mean()
        std_sales = sales_values.std()
        cv = (std_sales / avg_sales) if avg_sales > 0 else 0
        
        consistent_platforms.append({
            'Platform': platform,
            'Years_Active': years_active,
            'Total_Sales': total_sales,
            'Avg_Sales': avg_sales,
            'Std_Dev': std_sales,
            'CV': cv
        })

consistency_df = pd.DataFrame(consistent_platforms).sort_values('CV')

print("\nPlatforms with most consistent sales (Low Coefficient of Variation):")
print(consistency_df.head(10))
print("\n")

print("Most consistent platforms (CV < 0.5):")
most_consistent = consistency_df[consistency_df['CV'] < 0.5]
for _, row in most_consistent.head(5).iterrows():
    print(f"{row['Platform']}: {row['Years_Active']:.0f} years active, CV={row['CV']:.2f}")

print("\n" + "="*80 + "\n")

print("QUESTION 2: Can you identify platforms that have disappeared from the market?")
print("-" * 80)

disappeared_platforms = []

for platform in df['platform'].unique():
    platform_data = platform_yearly_sales[platform_yearly_sales['platform'] == platform].sort_values('year_of_release')
    
    if len(platform_data) > 0:
        first_year = platform_data['year_of_release'].min()
        last_year = platform_data['year_of_release'].max()
        peak_year = platform_data.loc[platform_data['total_sales'].idxmax(), 'year_of_release']
        peak_sales = platform_data['total_sales'].max()
        last_sales = platform_data[platform_data['year_of_release'] == last_year]['total_sales'].values[0]
        
        if last_year < 2015 and peak_sales > 20:
            disappeared_platforms.append({
                'Platform': platform,
                'First_Year': first_year,
                'Last_Year': last_year,
                'Peak_Year': peak_year,
                'Peak_Sales': peak_sales,
                'Last_Sales': last_sales,
                'Status': 'Disappeared'
            })

disappeared_df = pd.DataFrame(disappeared_platforms).sort_values('Peak_Sales', ascending=False)

print("\nPlatforms that have disappeared from the market:")
print(disappeared_df)
print("\n")

print("Notable disappeared platforms:")
for _, row in disappeared_df.head(10).iterrows():
    print(f"{row['Platform']}:")
    print(f"  Active: {row['First_Year']:.0f} - {row['Last_Year']:.0f}")
    print(f"  Peak: {row['Peak_Year']:.0f} with ${row['Peak_Sales']:.2f}M sales")
    print(f"  Last recorded: ${row['Last_Sales']:.2f}M in {row['Last_Year']:.0f}")
    print()

print("\n" + "="*80 + "\n")

print("QUESTION 3: What's the typical lifecycle of a gaming platform?")
print("-" * 80)

platform_lifecycles = []

for platform in df['platform'].unique():
    platform_data = platform_yearly_sales[platform_yearly_sales['platform'] == platform].sort_values('year_of_release')
    
    if len(platform_data) >= 3:
        first_year = platform_data['year_of_release'].min()
        last_year = platform_data['year_of_release'].max()
        lifecycle_length = last_year - first_year + 1
        
        peak_year = platform_data.loc[platform_data['total_sales'].idxmax(), 'year_of_release']
        time_to_peak = peak_year - first_year
        
        total_sales = platform_data['total_sales'].sum()
        
        platform_lifecycles.append({
            'Platform': platform,
            'First_Year': first_year,
            'Last_Year': last_year,
            'Lifecycle_Years': lifecycle_length,
            'Peak_Year': peak_year,
            'Years_to_Peak': time_to_peak,
            'Total_Sales': total_sales
        })

lifecycle_df = pd.DataFrame(platform_lifecycles)

print("\nPlatform lifecycle statistics:")
print(f"Average lifecycle: {lifecycle_df['Lifecycle_Years'].mean():.1f} years")
print(f"Median lifecycle: {lifecycle_df['Lifecycle_Years'].median():.1f} years")
print(f"Average time to peak: {lifecycle_df['Years_to_Peak'].mean():.1f} years")
print(f"Median time to peak: {lifecycle_df['Years_to_Peak'].median():.1f} years")
print("\n")

print("Lifecycle distribution:")
lifecycle_distribution = lifecycle_df['Lifecycle_Years'].value_counts().sort_index()
print(lifecycle_distribution.head(15))
print("\n")

print("Major platforms with longest lifecycles:")
long_lifecycle = lifecycle_df[lifecycle_df['Total_Sales'] > 100].sort_values('Lifecycle_Years', ascending=False).head(10)
print(long_lifecycle[['Platform', 'First_Year', 'Last_Year', 'Lifecycle_Years', 'Years_to_Peak', 'Total_Sales']])
print("\n")

print("LIFECYCLE PATTERNS:")
print("\n1. Introduction Phase (Years 1-2):")
print("   - Platform launches with limited game library")
print("   - Building user base and developer support")
print("   - Sales gradually increase")

print("\n2. Growth Phase (Years 2-4):")
print("   - Expanding game library")
print("   - Peak sales typically occur during this period")
print("   - Strong market presence")

print("\n3. Maturity Phase (Years 4-7):")
print("   - Market saturation")
print("   - Stable but declining sales")
print("   - Focus on established franchises")

print("\n4. Decline Phase (Years 7+):")
print("   - Next generation platforms emerge")
print("   - Developer support shifts to newer platforms")
print("   - Sales drop significantly")
print("   - Platform eventually discontinued")

print("\nTYPICAL LIFECYCLE CONCLUSION:")
print(f"- Most platforms have a lifecycle of 6-10 years")
print(f"- Peak sales occur around years 2-4 after launch")
print(f"- Major console generations (PS, Xbox, Nintendo) follow ~5-7 year cycles")
print(f"- Handheld systems may have longer lifecycles (8-10 years)")

print("\n" + "="*80 + "\n")

major_platforms = ['PS', 'PS2', 'PS3', 'PS4', 'X360', 'XOne', 'Wii', 'WiiU', 'DS', '3DS']
major_platform_data = lifecycle_df[lifecycle_df['Platform'].isin(major_platforms)].sort_values('First_Year')

plt.figure(figsize=(16, 8))
for _, row in major_platform_data.iterrows():
    plt.barh(row['Platform'], row['Lifecycle_Years'], 
             left=row['First_Year'], alpha=0.7, edgecolor='black')
    plt.plot(row['Peak_Year'], row['Platform'], 'r*', markersize=15)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Platform', fontsize=12)
plt.title('Platform Lifecycles Over Time (Red stars = Peak Sales Year)', fontsize=16, fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 6))
plt.hist(lifecycle_df['Lifecycle_Years'], bins=range(1, int(lifecycle_df['Lifecycle_Years'].max()) + 2), 
         color='steelblue', edgecolor='black', alpha=0.7)
plt.axvline(lifecycle_df['Lifecycle_Years'].mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Mean: {lifecycle_df["Lifecycle_Years"].mean():.1f} years')
plt.axvline(lifecycle_df['Lifecycle_Years'].median(), color='green', linestyle='--', 
            linewidth=2, label=f'Median: {lifecycle_df["Lifecycle_Years"].median():.1f} years')
plt.xlabel('Lifecycle Duration (Years)', fontsize=12)
plt.ylabel('Number of Platforms', fontsize=12)
plt.title('Distribution of Platform Lifecycle Lengths', fontsize=16, fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# Step 5 : Hypothesis Tests

‚ÄîAverage user ratings of the Xbox One and PC platforms are the same.

‚ÄîAverage user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

‚ÄîHow you formulated the null and alternative hypotheses

‚ÄîWhat criteria you used to test the hypotheses~~,~~ and why


```python
print("STEP 5: HYPOTHESIS TESTING")
print("\n" + "="*80 + "\n")

print("HYPOTHESIS TEST 1: Xbox One vs PC User Ratings")
print("="*80)

print("\n--- Formulating Hypotheses ---\n")

print("NULL HYPOTHESIS (H0):")
print("The average user ratings of Xbox One and PC platforms are the same.")
print("Mathematical notation: Œº(XOne) = Œº(PC)")
print("\n")

print("ALTERNATIVE HYPOTHESIS (H1):")
print("The average user ratings of Xbox One and PC platforms are different.")
print("Mathematical notation: Œº(XOne) ‚â† Œº(PC)")
print("\n")

print("TYPE OF TEST: Two-tailed test")
print("Reason: We're testing for any difference, not specifically whether one is higher or lower")
print("\n")

print("SIGNIFICANCE LEVEL (Alpha): Œ± = 0.05")
print("Reason: Standard significance level (5% risk of Type I error)")
print("This means we need 95% confidence to reject the null hypothesis")

print("\n" + "-"*80)
print("--- Data Preparation ---")
print("-"*80 + "\n")

xone_ratings = df_relevant[df_relevant['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = df_relevant[df_relevant['platform'] == 'PC']['user_score'].dropna()

print(f"Xbox One:")
print(f"  Sample size (n): {len(xone_ratings)}")
print(f"  Mean: {xone_ratings.mean():.4f}")
print(f"  Median: {xone_ratings.median():.4f}")
print(f"  Standard deviation: {xone_ratings.std():.4f}")
print(f"  Variance: {xone_ratings.var():.4f}")
print(f"  Min: {xone_ratings.min():.4f}")
print(f"  Max: {xone_ratings.max():.4f}")
print("\n")

print(f"PC:")
print(f"  Sample size (n): {len(pc_ratings)}")
print(f"  Mean: {pc_ratings.mean():.4f}")
print(f"  Median: {pc_ratings.median():.4f}")
print(f"  Standard deviation: {pc_ratings.std():.4f}")
print(f"  Variance: {pc_ratings.var():.4f}")
print(f"  Min: {pc_ratings.min():.4f}")
print(f"  Max: {pc_ratings.max():.4f}")
print("\n")

print(f"Difference in means: {abs(xone_ratings.mean() - pc_ratings.mean()):.4f}")

print("\n" + "-"*80)
print("--- Test Selection and Assumptions ---")
print("-"*80 + "\n")

print("STATISTICAL TEST: Independent Two-Sample t-Test (Welch's t-test)")
print("\n")

print("Criteria for test selection:")
print("1. INDEPENDENCE: ‚úì")
print("   - Xbox One and PC are different platforms")
print("   - User ratings are independent between platforms")
print("   - No paired or matched data")
print("\n")

print("2. MEASUREMENT LEVEL: ‚úì")
print("   - User scores are continuous numerical data (0-10 scale)")
print("   - Appropriate for parametric testing")
print("\n")

print("3. SAMPLE SIZE:")
print(f"   - Xbox One: {len(xone_ratings)} samples")
print(f"   - PC: {len(pc_ratings)} samples")
if len(xone_ratings) >= 30 and len(pc_ratings) >= 30:
    print("   - Both samples ‚â• 30: ‚úì (Central Limit Theorem applies)")
else:
    print("   - Sample sizes noted for interpretation")
print("\n")

print("4. VARIANCE ASSUMPTION:")
print("   - Using Welch's t-test (does NOT assume equal variances)")
print("   - More robust when variances differ between groups")
print("   - equal_var=False parameter in scipy.stats")

print("\n" + "-"*80)
print("--- Performing the Test ---")
print("-"*80 + "\n")

alpha_1 = 0.05

t_stat_1, p_value_1 = stats.ttest_ind(xone_ratings, pc_ratings, equal_var=False)

print(f"Test statistic (t): {t_stat_1:.4f}")
print(f"P-value: {p_value_1:.4f}")
print(f"Alpha level: {alpha_1}")
print(f"Degrees of freedom (approx): {len(xone_ratings) + len(pc_ratings) - 2}")

print("\n" + "-"*80)
print("--- Interpretation ---")
print("-"*80 + "\n")

print(f"Decision Rule:")
print(f"  If p-value < Œ± ({alpha_1}): Reject H0")
print(f"  If p-value ‚â• Œ± ({alpha_1}): Fail to reject H0")
print("\n")

print(f"Result: p-value ({p_value_1:.4f}) {'<' if p_value_1 < alpha_1 else '‚â•'} Œ± ({alpha_1})")
print("\n")

if p_value_1 < alpha_1:
    print("CONCLUSION: REJECT the null hypothesis")
    print(f"There IS a statistically significant difference between Xbox One and PC user ratings")
    print(f"Confidence level: {(1-alpha_1)*100}%")
    print(f"The difference of {abs(xone_ratings.mean() - pc_ratings.mean()):.4f} is statistically significant")
else:
    print("CONCLUSION: FAIL TO REJECT the null hypothesis")
    print(f"There is NO statistically significant difference between Xbox One and PC user ratings")
    print(f"The observed difference of {abs(xone_ratings.mean() - pc_ratings.mean()):.4f} could be due to random chance")

print("\n" + "="*80 + "\n")
print("\n")

print("HYPOTHESIS TEST 2: Action vs Sports Genre User Ratings")
print("="*80)

print("\n--- Formulating Hypotheses ---\n")

print("NULL HYPOTHESIS (H0):")
print("The average user ratings for Action and Sports genres are the same.")
print("Mathematical notation: Œº(Action) = Œº(Sports)")
print("\n")

print("ALTERNATIVE HYPOTHESIS (H1):")
print("The average user ratings for Action and Sports genres are different.")
print("Mathematical notation: Œº(Action) ‚â† Œº(Sports)")
print("\n")

print("TYPE OF TEST: Two-tailed test")
print("Reason: We're testing for any difference, not a specific directional hypothesis")
print("\n")

print("SIGNIFICANCE LEVEL (Alpha): Œ± = 0.05")
print("Reason: Consistent with Test 1, standard significance level")

print("\n" + "-"*80)
print("--- Data Preparation ---")
print("-"*80 + "\n")

action_ratings = df_relevant[df_relevant['genre'] == 'Action']['user_score'].dropna()
sports_ratings = df_relevant[df_relevant['genre'] == 'Sports']['user_score'].dropna()

print(f"Action Genre:")
print(f"  Sample size (n): {len(action_ratings)}")
print(f"  Mean: {action_ratings.mean():.4f}")
print(f"  Median: {action_ratings.median():.4f}")
print(f"  Standard deviation: {action_ratings.std():.4f}")
print(f"  Variance: {action_ratings.var():.4f}")
print(f"  Min: {action_ratings.min():.4f}")
print(f"  Max: {action_ratings.max():.4f}")
print("\n")

print(f"Sports Genre:")
print(f"  Sample size (n): {len(sports_ratings)}")
print(f"  Mean: {sports_ratings.mean():.4f}")
print(f"  Median: {sports_ratings.median():.4f}")
print(f"  Standard deviation: {sports_ratings.std():.4f}")
print(f"  Variance: {sports_ratings.var():.4f}")
print(f"  Min: {sports_ratings.min():.4f}")
print(f"  Max: {sports_ratings.max():.4f}")
print("\n")

print(f"Difference in means: {abs(action_ratings.mean() - sports_ratings.mean()):.4f}")

print("\n" + "-"*80)
print("--- Test Selection and Assumptions ---")
print("-"*80 + "\n")

print("STATISTICAL TEST: Independent Two-Sample t-Test (Welch's t-test)")
print("\n")

print("Criteria for test selection:")
print("1. INDEPENDENCE: ‚úì")
print("   - Action and Sports are different genres")
print("   - User ratings are independent between genres")
print("   - Games belong to distinct categories")
print("\n")

print("2. MEASUREMENT LEVEL: ‚úì")
print("   - User scores are continuous numerical data (0-10 scale)")
print("   - Appropriate for parametric testing")
print("\n")

print("3. SAMPLE SIZE:")
print(f"   - Action: {len(action_ratings)} samples")
print(f"   - Sports: {len(sports_ratings)} samples")
if len(action_ratings) >= 30 and len(sports_ratings) >= 30:
    print("   - Both samples ‚â• 30: ‚úì (Central Limit Theorem applies)")
    print("   - Large samples reduce impact of non-normality")
else:
    print("   - Sample sizes noted for interpretation")
print("\n")

print("4. VARIANCE ASSUMPTION:")
print("   - Using Welch's t-test (does NOT assume equal variances)")
print("   - Accounts for potential differences in rating variability")
print("   - More conservative and robust approach")

print("\n" + "-"*80)
print("--- Performing the Test ---")
print("-"*80 + "\n")

alpha_2 = 0.05

t_stat_2, p_value_2 = stats.ttest_ind(action_ratings, sports_ratings, equal_var=False)

print(f"Test statistic (t): {t_stat_2:.4f}")
print(f"P-value: {p_value_2:.4f}")
print(f"Alpha level: {alpha_2}")
print(f"Degrees of freedom (approx): {len(action_ratings) + len(sports_ratings) - 2}")

print("\n" + "-"*80)
print("--- Interpretation ---")
print("-"*80 + "\n")

print(f"Decision Rule:")
print(f"  If p-value < Œ± ({alpha_2}): Reject H0")
print(f"  If p-value ‚â• Œ± ({alpha_2}): Fail to reject H0")
print("\n")

print(f"Result: p-value ({p_value_2:.4f}) {'<' if p_value_2 < alpha_2 else '‚â•'} Œ± ({alpha_2})")
print("\n")

if p_value_2 < alpha_2:
    print("CONCLUSION: REJECT the null hypothesis")
    print(f"There IS a statistically significant difference between Action and Sports genre ratings")
    print(f"Confidence level: {(1-alpha_2)*100}%")
    print(f"The difference of {abs(action_ratings.mean() - sports_ratings.mean()):.4f} is statistically significant")
    if action_ratings.mean() > sports_ratings.mean():
        print(f"Action games are rated significantly higher than Sports games on average")
    else:
        print(f"Sports games are rated significantly higher than Action games on average")
else:
    print("CONCLUSION: FAIL TO REJECT the null hypothesis")
    print(f"There is NO statistically significant difference between Action and Sports genre ratings")
    print(f"The observed difference of {abs(action_ratings.mean() - sports_ratings.mean()):.4f} could be due to random chance")

print("\n" + "="*80 + "\n")

print("VISUAL COMPARISON OF DISTRIBUTIONS")
print("="*80 + "\n")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

axes[0, 0].hist(xone_ratings, bins=20, alpha=0.7, color='blue', edgecolor='black', label='Xbox One')
axes[0, 0].axvline(xone_ratings.mean(), color='darkblue', linestyle='--', linewidth=2, label=f'Mean: {xone_ratings.mean():.2f}')
axes[0, 0].set_title('Xbox One User Ratings Distribution', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('User Score', fontsize=10)
axes[0, 0].set_ylabel('Frequency', fontsize=10)
axes[0, 0].legend()
axes[0, 0].grid(axis='y', alpha=0.3)

axes[0, 1].hist(pc_ratings, bins=20, alpha=0.7, color='orange', edgecolor='black', label='PC')
axes[0, 1].axvline(pc_ratings.mean(), color='darkorange', linestyle='--', linewidth=2, label=f'Mean: {pc_ratings.mean():.2f}')
axes[0, 1].set_title('PC User Ratings Distribution', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('User Score', fontsize=10)
axes[0, 1].set_ylabel('Frequency', fontsize=10)
axes[0, 1].legend()
axes[0, 1].grid(axis='y', alpha=0.3)

axes[1, 0].hist(action_ratings, bins=20, alpha=0.7, color='green', edgecolor='black', label='Action')
axes[1, 0].axvline(action_ratings.mean(), color='darkgreen', linestyle='--', linewidth=2, label=f'Mean: {action_ratings.mean():.2f}')
axes[1, 0].set_title('Action Genre User Ratings Distribution', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('User Score', fontsize=10)
axes[1, 0].set_ylabel('Frequency', fontsize=10)
axes[1, 0].legend()
axes[1, 0].grid(axis='y', alpha=0.3)

axes[1, 1].hist(sports_ratings, bins=20, alpha=0.7, color='red', edgecolor='black', label='Sports')
axes[1, 1].axvline(sports_ratings.mean(), color='darkred', linestyle='--', linewidth=2, label=f'Mean: {sports_ratings.mean():.2f}')
axes[1, 1].set_title('Sports Genre User Ratings Distribution', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('User Score', fontsize=10)
axes[1, 1].set_ylabel('Frequency', fontsize=10)
axes[1, 1].legend()
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

box_data_1 = [xone_ratings, pc_ratings]
axes[0].boxplot(box_data_1, labels=['Xbox One', 'PC'], patch_artist=True,
                boxprops=dict(facecolor='lightblue', alpha=0.7),
                medianprops=dict(color='red', linewidth=2))
axes[0].set_title('User Ratings Comparison: Xbox One vs PC', fontsize=12, fontweight='bold')
axes[0].set_ylabel('User Score', fontsize=10)
axes[0].grid(axis='y', alpha=0.3)
axes[0].text(1.5, axes[0].get_ylim()[1]*0.95, f'p-value: {p_value_1:.4f}', 
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

box_data_2 = [action_ratings, sports_ratings]
axes[1].boxplot(box_data_2, labels=['Action', 'Sports'], patch_artist=True,
                boxprops=dict(facecolor='lightgreen', alpha=0.7),
                medianprops=dict(color='red', linewidth=2))
axes[1].set_title('User Ratings Comparison: Action vs Sports', fontsize=12, fontweight='bold')
axes[1].set_ylabel('User Score', fontsize=10)
axes[1].grid(axis='y', alpha=0.3)
axes[1].text(1.5, axes[1].get_ylim()[1]*0.95, f'p-value: {p_value_2:.4f}', 
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("SUMMARY OF HYPOTHESIS TESTS")
print("="*80 + "\n")

summary_data = {
    'Test': ['Xbox One vs PC', 'Action vs Sports'],
    'H0': ['Œº(XOne) = Œº(PC)', 'Œº(Action) = Œº(Sports)'],
    'H1': ['Œº(XOne) ‚â† Œº(PC)', 'Œº(Action) ‚â† Œº(Sports)'],
    'Test_Used': ['Welch t-test', 'Welch t-test'],
    'Alpha': [alpha_1, alpha_2],
    't_statistic': [t_stat_1, t_stat_2],
    'p_value': [p_value_1, p_value_2],
    'Decision': ['Reject H0' if p_value_1 < alpha_1 else 'Fail to reject H0',
                 'Reject H0' if p_value_2 < alpha_2 else 'Fail to reject H0']
}

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

print("\n")
print("KEY TAKEAWAYS:")
print("\n1. Both tests used Welch's t-test (independent samples, unequal variances)")
print("2. Alpha level of 0.05 provides 95% confidence in results")
print("3. Two-tailed tests appropriate for testing 'difference' (not direction)")
print("4. Large sample sizes support validity of Central Limit Theorem")
print("5. Results provide statistical evidence for decision-making in marketing strategy")
```
